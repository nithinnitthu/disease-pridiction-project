```python
import pandas as pd
```

```python
data = pd.read_csv('Downloads/Disease_train.csv')
```

```python
data
```

```python
def fill_missing_value(featur):
    mean = featur.mean()
    return featur.fillna(mean)
```

```python
for i in range(10):
    data[f"feature_{i+1}"] = fill_missing_value(data[f"feature_{i+1}"])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X = data.drop(['patient_id','diagnosis'],axis=1)
```

```python
y = data['diagnosis']
```

```python
X_stand = sc.fit_transform(X)
```

```python
X_stand
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(y.value_counts())
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_stand,y,test_size=0.25,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smot = SMOTE()
```

```python
X_train_rs,y_train_rs = smot.fit_resample(X_train,y_train)
```

```python
X_train_rs
```

```python
y_train_rs
```

```python
print(y_train_rs.value_counts())
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train_rs, y_train_rs)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
}
```

```python
GridCV = GridSearchCV(estimator=svc, param_grid=parameters, scoring='roc_auc', cv=5, verbose=2, n_jobs=-1)
```

```python
GridCV.fit(X_train_rs, y_train_rs)
```

```python
best_svc = GridCV.best_estimator_
```

```python
y_predicted = best_svc.predict(X_test)
```

```python
print(accuracy_score(y_predicted,y_test))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_predicted,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_predicted,y_test))
```

```python
print(accuracy_score(y_predicted,y_test))
```

```python
print(roc_auc_score(y_predicted,y_test))
```
